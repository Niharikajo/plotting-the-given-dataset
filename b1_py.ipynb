```python
#importing numpy pandas libraries 
import numpy as np 
import pandas as pd
```

```python
#read saved a csv file
data=pd.read_csv('batterydata.csv')
```

```python
#print the loaded csv file
data
```

```python
#importing matplotlib.pyplot library to plot graphs 
import matplotlib.pyplot as plt
```

```python
#sctter plot of time vs battery level
plt.scatter(data['Time'],data['Battery_Level']) #time is taken in the x axis and battery level in y axis
#labeling thr axes
plt.xlabel('time')
plt.ylabel('battery level')
#title of the plot
plt.title('scatter plot time vs battery level')
plt.xticks(rotation=60) # rotates x axis ticks values by 60
```

```python
#line graph of time vs battery level
plt.plot(data['Time'],data['Battery_Level'])
# giving labels to x and y axis
plt.xlabel('time')
plt.ylabel('battery level and temperature')
plt.title('line plot time vs battery level') #title of the plot 
plt.xticks(rotation=50)
plt.xticks(data.Time[::2]) # skips every two ticks of x axis
```

```python
#line plot of time vs battery level and time vs temperature in a single plot with only 2 axes
plt.plot(data['Time'],data['Battery_Level']) # plots time vs batterylevel
plt.plot(data['Time'],data['Temperature']) #plots time vs temperature
#labeling x and y axis
plt.xlabel('time')
plt.ylabel('battery level and temperature')
plt.title('plot time vs battery level and time vs temperature') # title of the plot
plt.legend(['BL','temp']) #we have two plots in the same graph so we use legends identify different plots
plt.xticks(rotation=50)
plt.xticks(data.Time[::3]) #skips every three values in between in x axis
```

```python
# two plots in a single graph with twin axis
fig,ax1=plt.subplots() #subplot is added
color='tab:orange'
ax1.set_xlabel('Time') # x axis label 
ax1.set_ylabel('Battery Level',color=color) #y axis label and color is set
ax1.plot(data['Time'],data['Battery_Level'],color=color,label='bl') #ploting time vs battery level
ax1.legend(loc='best') # legend for battery level with location specified as best
plt.xticks(rotation=60)
ax2=ax1.twinx() # twin axis created 
color='tab:cyan'
ax2.set_ylabel('temperature',color=color) # secondary y axis label and color of the label is set 
ax2.plot(data['Time'],data['Temperature'],color=color,label='temp') # time vs temperature is plotted
ax2.legend(loc='upper left',bbox_to_anchor=(0.0,0.9)) #legend for temperature with specified location
plt.xticks(data.Time[::3]) 
fig.suptitle('Time vs Battery Level Time vs Temparature') #title of the plot
plt.show() #complete plot is displayed

```

```python
#plot of Time vs Voltage vs Current using twin axis 
fig,ax1=plt.subplots() # subplot is created
color='tab:red'
ax1.set_xlabel('Time') # x axis labelis set
ax1.set_ylabel('Battery Level',color=color) # y axis label and label color is set
ax1.plot(data['Time'],data['Battery_Level'],color=color,label='battery level') #plot time vs battery level
ax1.legend(loc=(0.0,0.7)) # here we use two different legends (for time vs battery level and time vs current) 
#legend with specified location for time vs battery level loctaion
plt.xticks(rotation=60)
ax2=ax1.twinx() # twin axis is created
color='tab:blue'
ax2.set_ylabel('Current',color=color) # twin y axis label and label color is set
ax2.plot(data['Time'],data['Current'],color=color,label='current') # second plot of time vs current is plotted
ax2.legend(loc=(0.01,0.8)) #location of second legend is specified
plt.xticks(data.Time[::3])
fig.suptitle('Time vs Voltage vs Current') #title of the plot
plt.show() # complete plot is displayed
```

```python
#plot of time vs battery level vs current using twin axis and single legend for both plots 
fig,ax1=plt.subplots() # subplot is created
color='tab:red'
ax1.set_xlabel('Time') # x axis label is set
ax1.set_ylabel('Battery Level',color=color)# y axis label and label color is set
ax1.tick_params(axis='y',colors='red') # y axis tick color is set as red
l1=ax1.plot(data['Time'],data['Battery_Level'],color=color,label='battery level') # first plot of time vs battery level
plt.xticks(rotation=60) # x ticks are rotated by 60
ax2=ax1.twinx() # twin axis is created
color='tab:blue'
ax2.set_ylabel('Current',color=color) # twin y axis label and label color is set
l2=ax2.plot(data['Time'],data['Current'],color=color,label='Current') # second plot of time vs current
plt.xticks(data.Time[::3]) # skips 3 vales in between (x axis)
ax2.tick_params(axis='y',colors='blue') # twin y axis label color is set as blue
#single label for both the plots
lines = l1 +l2
labels = [l.get_label() for l in lines]
ax1.legend(lines,labels,loc='lower center')
fig.suptitle('Time vs Battery level vs Current',color='green') # title of the plot
plt.tight_layout() # automatically adjusts subplot params
plt.show()# displayes complete plot
```
